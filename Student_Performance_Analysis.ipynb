## Internship Task 1 - Data Analysis with Python

This project analyzes student performance data using Python. The workflow includes data loading, cleaning, analysis, visualization, and drawing conclusions.

```python
import os

print(os.listdir())
['student-mat.csv']
```

## Dataset Loading

The Student Performance dataset was loaded using the Pandas library. The first few rows were displayed to verify that the dataset was imported successfully.

```python
import pandas as pd

df = pd.read_csv('student-mat.csv', sep=';')

df.head()
```

## Data Exploration

The dataset shape, column names, and data types were examined to understand the structure of the data.

```python
df.shape
```

## Data Cleaning

The dataset was checked for missing values and duplicate records. Necessary cleaning steps were performed to ensure data quality before analysis.

```python
df.isnull().sum()
```

## Average Final Grade (G3)



The average final grade of all students is **10.42**.

```python
df['G3'].mean()
```

## Students Scoring Above 15

A total of **40** students scored above 15 in the final grade (G3).

```python
len(df[df['G3'] > 15])
```

## Study Time and Performance

The correlation between study time and final grade is **0.0978**, indicating a weak positive relationship. Students who spend more time studying tend to achieve slightly better grades, although the effect is not strong.

```python
df['studytime'].corr(df['G3'])
```


## Gender-wise Performance Analysis

The average final grades (G3) were compared between male and female students.

| Gender | Average Final Grade (G3) |
|---------|-------------------------|
| Female | 9.97 |
| Male | 10.91 |

The analysis shows that **male students performed better on average**, achieving an average final grade of **10.91**, compared to **9.97** for female students.

This difference suggests that male students in this dataset had slightly higher academic performance in terms of final grades.

```python
df.groupby('sex')['G3'].mean()
```

## Grade Distribution

The histogram shows the distribution of final grades among students. Most students scored within the middle range, while fewer students achieved very high or very low grades.

```python
import matplotlib.pyplot as plt

plt.hist(df['G3'])
plt.title('Distribution of Final Grades')
plt.xlabel('G3')
plt.ylabel('Count')
plt.show()
```

## Study Time vs Final Grade

The scatter plot visualizes the relationship between study time and final grades. It helps identify whether increased study time is associated with better academic performance.

```python
plt.scatter(df['studytime'], df['G3'])
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time')
plt.ylabel('G3')
plt.show()
```

## Gender Comparison

The bar chart compares the average final grades of male and female students, providing a clear visual comparison of performance.

```python
gender_avg = df.groupby('sex')['G3'].mean()

gender_avg.plot(kind='bar')

plt.title('Average Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average G3')
plt.show()
```
